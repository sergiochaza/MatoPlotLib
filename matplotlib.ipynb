#Modo Api

Llamamos a la api mediante PLT, para poder realizar los graficos

```python
#Metodo plt, API
plt.figure(figsize=(10,8))
plt.title("Mi Primer Grafico")
plt.plot(x, x**2)
plt.plot(x, -1 * (x**2))
plt.show()
```

```python
#segundo grafico Modo API: plt

plt.figure(figsize=(12,6))
plt.title("Mi Segundo Grafico")

plt.subplot(1,2,1) #Fila, Columna, Panel seleccionado
plt.plot(x, x**2)
plt.plot([0,0,0], [-10,0,100])
plt.legend(["x**2", "Linea Vertical"])
plt.xlabel("X")
plt.ylabel("X al cuadrado")

plt.subplot(1,2,2)
plt.plot(x, -1 * (x**2))
plt.plot([-10,0,10], [-50,-50,-50])
plt.legend(["-x**2", "Linea Horizontal"])
plt.xlabel("X")
plt.ylabel("-X al Squared")

plt.xlabel("X")
plt.ylabel("X al cuadrado")
plt.show()
```

#Grafico de barras


```python
import pandas as pd
paises_latam = [
    "Argentina", "Bolivia", "Brasil", "Chile", "Colombia",
    "Costa Rica", "Cuba", "Ecuador", "El Salvador", "Guatemala",
    "Honduras", "México", "Nicaragua", "Panamá", "Paraguay",
    "Perú", "República Dominicana", "Uruguay", "Venezuela"
]

poblacion_millones = [
    46.2, 12.3, 215.3, 19.5, 52.2,
    5.3, 11.1, 17.9, 6.5, 18.3,
    10.4, 129.9, 6.8, 4.5, 7.5,
    34.6, 11.2, 3.5, 28.1
]
pobla_paises = pd.DataFrame({"pais": paises_latam, "poblacion_mill": poblacion_millones})
pobla_paises.head()
```

```python
pobla_paises = pobla_paises.sort_values(by="poblacion_mill", ascending=False)
```

```python
#Grafico de Barras
plt.figure(figsize=(20,6))
import matplotlib.pyplot as plt
plt.bar(pobla_paises["pais"], pobla_paises["poblacion_mill"], data=pobla_paises, color="red")
plt.title("Población Latinoamericana")
plt.xlabel("Paises")
plt.xticks(rotation=50)
plt.ylabel("Población en millones")
plt.grid(True)

for indice, valor in enumerate(pobla_paises["poblacion_mill"]):
    plt.text(indice, valor+1, str(valor), ha="center")
plt.show()
```

#Grafico de Tortas
Como se distribuyen las partes de un conjunto total

```python
import pandas as pd
import matplotlib.pyplot as plt
datos ={
    "marca": [
        "Toyota", "Volkswagen", "Hyundai", "General Motors", "Ford",
        "Honda", "Nissan", "BMW", "Mercedes-Benz", "Kia"
    ],
    "unidades_vendidas_millones": [
        10.4, 8.9, 6.8, 6.3, 5.9,
        4.3, 4.1, 2.5, 2.3, 2.0
    ]
}


df = pd.DataFrame(datos)
df.head()
```

```python
#Grafico de torta o pie
plt.figure(figsize=(20,8))
plt.pie(df["unidades_vendidas_millones"], labels=df["marca"], autopct="%1.1f%%")
plt.title("Unidades Vendidas por Marca")
plt.show()
```

```python
#Grafico de torta o pie
colores_pie = [
    "#E69F00",  # naranja suave
    "#56B4E9",  # azul cielo
    "#009E73",  # verde
    "#F0E442",  # amarillo
    "#0072B2",  # azul más profundo
    "#D55E00",  # rojo anaranjado
    "#CC79A7",  # rosa/magenta
    "#999999",  # gris neutro
    "#66C2A5",  # verde agua
    "#FC8D62"   # coral pastel
]
plt.figure(figsize=(20,8))
plt.pie(
    df["unidades_vendidas_millones"],
    labels=df["marca"],
    autopct="%1.1f%%",
    startangle=90, colors=colores_pie,
    textprops={"fontsize": 10})
plt.title("Unidades Vendidas por Marca")
plt.show()
```

#Histogramas

Agarra todos los valores del eje x y lo separa en 50 partes en este caso y la altura de cada parte es la cantidad de valores que aparecen en tiempode respuesta dentro del rango


las barras mas altas tienen mas valores dentro del bins

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Tiempo de respuesta de un servidor web
tiempos_respuestas = np.random.normal(loc=200, scale=50, size=1000) #Media=200ms, desv, normal es campana de gaus
```

```python
#Grafico histograma
plt.figure(figsize=(10,6))
plt.hist(tiempos_respuestas, bins=20, alpha=0.7) #Bins número de cuadros
plt.show()
```

```python
#Grafico histograma
plt.figure(figsize=(10,6))
plt.hist(tiempos_respuestas, bins=100, alpha=0.7, density=True) #Bins número de cuadros, calcula la probabilidad
plt.show()
```

#Graficos de Dispersión o de puntos (scatterplot)


```python
np.random.seed(42)
#Simular datos de horas de estudio y calificación en examen

horas_estudio = np.random.randint(1,11,100) #Horas de estudio de 1 a 10 horas
calificaciones = horas_estudio * 10 + np.random.randint(0,5,100) #Relacion lineal + ruido
```

```python
df = pd.DataFrame({"horas_estudio": horas_estudio, "calificaciones": calificaciones})
df.head()
```

```python
#Grafico de dispersión o puntos: plt.scatter
plt.figure(figsize=(10,6))
plt.scatter(df["horas_estudio"], df["calificaciones"], alpha=0.5) #Alpha bajo es la transparencia para valores que se repiten, ver en un mapa de calor cuales son los puntos que se repiten mas, un valor que aparece una unica vez o veo bajito y uno que se repite mucho bastante
plt.grid(True)
plt.show()
```
